### Import the Required Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import silhouette_score
import scipy.cluster.hierarchy as sch 
```

### Read Given Dataset

```python
# Importing dataset by pandas
data = pd.read_csv('../datasets/Country_clusters.csv')
```

```python
df = data.loc[:, ["Lattitude","Longitude"]]
df
```

### Import cluster hierarchy.

```python
plt.figure(figsize=(5, 3))
plt.scatter(df[["Lattitude"]], df[["Longitude"]], s=50, c='blue')
```

### Plot dendrogram

```python
plt.figure(figsize=(6,4))


dendrogram = sch.dendrogram(sch.linkage(df, method  = "complete")) #single,complete,average,ward
plt.title('Dendrogram')
plt.xlabel('Lattitude')
plt.ylabel('Longitude')
plt.show()
```
