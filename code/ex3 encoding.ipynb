#### Encode 'Species' column by using different encoding techniques.
1.  Simple Encoding
2.  Label encoding
3.  One hot encoding
4.  Get dummies  

#### 1. Simple Encoding

```python
df['Species'].unique()
```

```python
df['Species']=df['Species'].map({'setosa':0, 'versicolor':1, 'virginica':2})
```

```python
df.head()
```

#### 2. Label Encoding

```python
df = pd.read_csv("../datasets/Iris.csv")
```

```python
label_enco = LabelEncoder()
```

```python
df['Species']= label_enco.fit_transform(df['Species'])
```

```python
df.head()
```

#### 3. One Hot Encoding 

```python
df = pd.read_csv("../datasets/Iris.csv")
```

```python
one_hot_encoder = OneHotEncoder(sparse_output=False)
```

```python
encoded = one_hot_encoder.fit_transform(df[['Species']]) 
```

```python
# Concat Data Frame to Main Data
```

```python
encoded_df = pd.DataFrame(
    data = encoded,
    columns=one_hot_encoder.get_feature_names_out(['Species'])
)
```

```python
pd.concat([df.drop(['Species'],axis=1),encoded_df],axis=1)
```

#### 4. Get dummies 

```python
df = pd.read_csv("../datasets/Iris.csv")
```

```python
species_dummy= pd.get_dummies(df['Species'],drop_first=True,dtype=int)
```

```python
df= df.drop(['Species'],axis=1)
```

```python
pd.concat([df,species_dummy],axis=1)
```
