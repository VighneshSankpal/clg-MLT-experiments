### Fill the null values with statistical method 

```python
df.isna().sum()
```

```python
msno.matrix(df,figsize=(6,3),color=(0.4, 0.2, 0.6),sparkline=False);
```

```python
df.info()
```

```python
plt.hist(df['Age']);
plt.xlabel('Age',fontsize=12);
```

#### In Age column data is not normally distributed. 
#### Theirfor we use Median value to fill null values.

```python
df['Age']=df['Age'].fillna(value=df['Age'].median())
```

```python
msno.matrix(df,figsize=(6,3),color=(0.4, 0.2, 0.6),sparkline=False);
```
